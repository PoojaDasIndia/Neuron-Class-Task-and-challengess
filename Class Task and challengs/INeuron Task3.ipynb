### Question 6. Try to extract all the numerical data it may be a part of dict key and values

```python
list_of_numeric=[]
for i in l:
    for j in i:
        if type(j)==int :
            list_of_numeric.append(j)
    if type(i)==dict:
        for k in i:
            if type(k)==int or type(i[k])== int:
                list_of_numeric.append(k)
                list_of_numeric.append(i[k])
print(list_of_numeric)                
```

### Question 15. Try to unwrape all the collection inside collection and create a flat list


```python
All_element_list=[]
for i in l:
    if type(i)!=dict:
        for j in i:
            if type(j)!=dict:
                All_element_list.append(j)
            if type(j)==dict:
                for k in j:
                    All_element_list.append(j[k])
            
    if type(i)==dict:
        for j in i:
            All_element_list.append(i[j])
print(All_element_list)       
```

### Question 10.  Try to find out a number of occurances of all the data

```python
print("Number  of  element occurance \n\n")
for i in set(All_element_list):
    print(i," : ",All_element_list.count(i))
```
